## Check Normal Distribution Using QQ Plot-Quantile â€“ Quantile plot 

```python
### importing necessary modules
import scipy.stats as stat
import pylab   ## help in displaying plots from matplotlib.pyplot
import numpy as np
```

```python
###  ## library for handling data and it has datasets too
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
### histogram withous smoothning


#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.grid()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)  ## pylab used
    plt.show()
    
```

```python
plot_data(df,'sepal_width')
```

```python
## smoothen histogram

#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
#     plt.grid()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)  ## pylab used
    plt.show()
    
```

```python
plot_data(df,'sepal_width')
```

```python
plot_data(df,'petal_width')
```

```python
plot_data(df,'sepal_length')
```

```python
plot_data(df,'petal_length')
```
