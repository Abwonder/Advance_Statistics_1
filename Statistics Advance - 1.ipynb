Q1. What is the Probability density function?

The probability density function (PDF) is a mathematical function that describes the probability distribution of a continuous random variable. The PDF provides the probability of a random variable taking on a certain value or falling within a certain range of values, and it is typically represented graphically as a curve.

The PDF of a continuous random variable has the property that the area under the curve between two points on the x-axis represents the probability that the random variable falls within that range of values. In other words, the integral of the PDF over a certain interval gives the probability of the random variable falling within that interval.


### Question 2

Q2. What are the types of Probability distribution?

__Types of Probability Distribution__

There are several types of probability distributions, including:

__Probability density Function (Discrete probability distribution)__

This type of distribution is used to model the probability of discrete or countable outcomes, such as the number of heads in a series of coin flips. Examples of discrete probability distributions include the binomial distribution, the Poisson distribution, and the geometric distribution.

__Probability mass Function (continuous probability distribution)__

This type of distribution is used to model the probability of continuous outcomes, such as the height or weight of a population. Examples of continuous probability distributions include the normal distribution, the exponential distribution, and the beta distribution.

__Bernoulli distribution__

This is a special case of the binomial distribution where there is only one trial, and the outcome is either success or failure.


__Binomial distribution__ 

This distribution is used to model the probability of a certain number of successes in a fixed number of independent trials, each with the same probability of success.

__Poisson distribution__

This distribution is used to model the probability of a certain number of events occurring in a fixed time or space interval, when the events occur randomly and independently at a constant rate.

__Normal distribution__

Also known as the Gaussian distribution, this is a continuous probability distribution that is commonly used to model the distribution of many natural phenomena, such as heights, weights, and test scores. It has a bell-shaped curve and is characterized by its mean and standard deviation.

__Exponential distribution__

This is a continuous probability distribution that is commonly used to model the time between two successive events in a Poisson process, such as the time between car accidents on a particular stretch of road.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.



```python
### import library
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters of the binomial distribution
n = 10   # number of trials
p = 0.4  # probability of success

# Generate a random sample of size 1000 from the binomial distribution
sample = np.random.binomial(n, p, size=1000)

# Plot a histogram of the sample using matplotlib
plt.hist(sample, bins=range(n+1))
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n={}, p={})".format(n, p))
plt.show()
```

```python

import scipy.stats as stats
## function to calculate e the cumulative distribution function of a Poisson distribution
# with given mean at a given point.
def poisson_cdf(mean, point):
    return stats.poisson.cdf(point, mu=mean)

# mean: the mean of the Poisson distribution
# point: the point at which to evaluate the CDF
```

### Question 7

Q7. How Binomial distribution different from Poisson distribution?

Differences in binomial distribution and poisson distribution can be based on the following.

Number of trials: In a binomial distribution, the number of trials is fixed and finite, while in a Poisson distribution, the number of events can be infinite and is not fixed in advance.

Probability of success: In a binomial distribution, the probability of success is constant across all trials, while in a Poisson distribution, the probability of an event occurring is constant across a fixed time interval, but can vary across different time intervals.

Relationship between mean and variance: In a binomial distribution, the mean and variance are related by the formula mean = n * p and variance = n * p * (1 - p), where n is the number of trials and p is the probability of success. In a Poisson distribution, the mean and variance are equal, and are both equal to the parameter λ that represents the average rate of events per unit of time or space.

Shape of distribution: In a binomial distribution, the distribution is typically symmetric and bell-shaped, while in a Poisson distribution, the distribution is typically skewed to the right and has a longer tail on the right side.


### Question 8

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# set the random seed for reproducibility
np.random.seed(123)

# generate a random sample of size 1000 from a Poisson distribution with mean 5
sample = np.random.poisson(5, size=1000)

# calculate the sample mean and variance
sample_mean = np.mean(sample)
sample_var = np.var(sample, ddof=1)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_var)
```

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

With respect to mean position, the data that are away farther away from the mean has less frequency, so in normal distribution, the least frequent data appears in the tails of the distribution, which are located at the extremes of the range of values. The tails are the areas of the distribution that fall beyond a certain number of standard deviations from the mean.

__Using standard deviation to buttress this point:__

Specifically, in a normal distribution, approximately 68% of the data falls within one standard deviation of the mean, approximately 95% falls within two standard deviations, and approximately 99.7% falls within three standard deviations. This leaves only a small percentage of the data, less than 0.3%, in the tails of the distribution beyond three standard deviations from the mean. Therefore, the data that appears in the tails of the distribution is much less frequent than the data that appears closer to the mean.
